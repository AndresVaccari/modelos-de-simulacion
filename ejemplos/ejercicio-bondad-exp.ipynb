```python
from scipy.stats import binom, poisson, norm, expon
import numpy as np
import matplotlib.pyplot as plt
```

```python
sample = [ 3,	15,	45,	3,	8,	11,	2,	23,	21,	5, 13,	2,	1,	5,	16,	9,	3,	16,	10,	13]
beta= 10

x = np.arange(expon.ppf(0.01, scale=beta),
              expon.ppf(0.99, scale=beta))

data_expon = expon.pdf(x, scale=beta)
_ = plt.plot(x, data_expon, 'r-')
_ = plt.hist(sample, density=1, bins=4)
plt.show()

#print(data_expon)
histo = np.histogram(sample,bins=4)
print (histo)
print(histo[0])
print(histo[1])
prob = expon.cdf(histo[1][1:5], scale=beta)
print("ACC Expon Intervalos = {m}".format(m=prob))

intervals_histo = np.diff(prob, n=1)
print("Diff Intervalos = {m}".format(m=intervals_histo))
mergedList = np.array([prob[0]])
#mergedList.extend(intervals_histo)
mergedList = np.append(mergedList, intervals_histo)

print("Merged List = {m}".format(m=mergedList))
frecuencias = mergedList * len(sample)
print("Frecuencias = {m}".format(m=frecuencias))

```
